```python
import bs4
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import time
```

```python
c=[]
```

```python
def stock():
    words=["As of  ","EDT."]
    hdr = {'User-Agent': 'Mozilla/5.0'}
    url= requests.get("https://finance.yahoo.com/quote/FAST?p=FAST&.tsrc=fin-tre-srch",headers=hdr)
    soup=bs4.BeautifulSoup(url.text,"html")
    body=soup.findAll("div",{"class":"My(6px) Pos(r) smartphone_Mt(6px)"})
    value=body[0].findAll("span")
    return dict(stock=float(value[0].text),
    incresepercent=str(value[1].text),
    time=str(value[2].text[:value[2].text.find("EDT")].replace("As of  ","")),
    market_status=str(value[2].text[value[2].text.find("Market"):]))
```

```python
t=time.time()
while True:
    if time.time()-t>2:
        #b=list(stock(body).values())
        b=stock()
        c.append(b)
        df2=pd.DataFrame(c)
        print(df2)
        t=time.time()
        
```
